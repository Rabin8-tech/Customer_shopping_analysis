```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

```python
df.isnull().sum()
```

```python
labels=['Young Adult','Adult','Middle-Aged','Senior-Citizen']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
frequency_mapping={'Fortnightly':14,'Weekly':7,'Monthly':30,'Quarterly':90,'Bi-Weekly':14,'Annually':365,'Every 3 Months':90}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
import mysql.connector
import pandas as pd

# --- Fix categorical columns ---
df = df.apply(lambda x: x.astype(str) if x.dtype.name == 'category' else x)
df = df.fillna("").astype(str)

# --- Connect to MySQL ---
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="rabin8866",
    database="data_project"
)
cursor = conn.cursor()

# --- Create table dynamically ---
table_name = "customer"
columns = df.columns
create_table_query = f"CREATE TABLE IF NOT EXISTS {table_name} ("
for col in columns:
    create_table_query += f"`{col}` VARCHAR(255),"
create_table_query = create_table_query.rstrip(',') + ")"
cursor.execute(create_table_query)

# --- Insert data ---
cols = ",".join([f"`{col}`" for col in df.columns])
placeholders = ",".join(["%s"] * len(df.columns))
insert_query = f"INSERT INTO {table_name} ({cols}) VALUES ({placeholders})"
data = [tuple(x) for x in df.to_numpy()]
cursor.executemany(insert_query, data)
conn.commit()

print(f"{cursor.rowcount} rows inserted successfully into '{table_name}'!")

cursor.close()
conn.close()

```
